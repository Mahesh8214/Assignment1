```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Note: Also add title, xlabel, and ylabel to the plot.

# Ans: 
    
#     A scatter plot is a type of plot that displays individual data points as markers on a two-dimensional plane.
#     It is commonly used to visualize the relationship or correlation between two continuous variables.


plt.ylabel('y values')
plt.xlabel('x values')
plt.title('scatter plot of random data');
plt.scatter(x,y)

```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

# Ans: 
#     The subplot() function in Matplotlib is used to create multiple plots in a single figure, arranged in a grid-like structure. 
#     It allows you to divide the figure into multiple subplots, specifying the number of rows, columns, and the index of the current subplot.
    
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()

    
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Ans:
     # A bar plot, also known as a bar chart, is a type of plot that uses rectangular bars to represent categorical data.
     # It displays the values of different categories or groups as vertical or horizontal bars, 
     # with the length or height of each bar proportional to the corresponding value.
plt.barh(company , profit) 
    
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
 

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Values")

plt.show()
```
